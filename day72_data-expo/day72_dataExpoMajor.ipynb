```python
# we did find a NaN entry at 50, so we can remove this using .dropna()

clean_df = df.dropna() # removes missing values
clean_df.tail()
```

```python
# to find the max of it:
clean_df['Starting Median Salary'].max()

# to find the id of the max value we use:
clean_df['Starting Median Salary'].idxmax()
```

```python
# To see the name of the major that corresponds to
# that particular row, we can use the .loc[]
clean_df['Undergraduate Major'].loc[43]

# to print values of all the columns for this perticular row:
clean_df.loc[43]
```

```python
# CHALLENGE

# What college major has the highest mid-career salary?
# How much do graduates with this major earn?
clean_df['Mid-Career Median Salary'].idxmax()
# -> 8
# clean_df['Undergraduate Major'].loc[8]
# clean_df['Undergraduate Major'][8]
clean_df.loc[8]
```

```python
# Which college major has the lowest starting salary and
# how much do graduates earn after university?
clean_df['Starting Median Salary'].idxmin()
# clean_df['Undergraduate Major'].loc[49]
clean_df.loc[49]
```

```python
# Which college major has the lowest mid-career salary and
#  how much can people expect to earn with this degree?
clean_df['Mid-Career Median Salary'].idxmin()
clean_df['Undergraduate Major'].loc[18]
clean_df.loc[18]
```

```python
# calculate the difference between the earnings of the 10th and 90th percentile:
# it is also known as the spread
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']

# we can insert this column in our dataframe using insert():
# clean_df.insert(1, 'Spread', spread_col)
# (index position, name of column, the column to be added)
clean_df.head()
```

```python
# SORTING:

low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
# sorting the column undergraduate major, w.r.t Spread, and displaying top 5 values
```

```python
# find the degrees with the highest potential?

# Find the top 5 degrees with the highest values in the 90th percentile.
highest_pot = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_pot[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()

```

```python
# GROUPING AND PIVOTING DATA:

clean_df.groupby('Group').count()
# count() give us the number of entries in that catagory

clean_df.groupby('Group').mean()

pd.options.display.float_format = '{:,.2f}'.format

clean_df.groupby('Group').mean()
```
